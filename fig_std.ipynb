```python
import pyFAI,numpy
from pyFAI.worker import Worker
```

```python
ai = pyFAI.load({"detector":"pilatus100k", "wavelength":1e-10})
ai
```

```python
img = numpy.ones(ai.detector.shape)
img.shape
```

```python
ai.integrate1d(img, 100, variance=-img, error_model="poisson", unit="r_mm",radial_range=(0,100))
```

```python
worker = Worker(ai, shapeOut=(1,100), method=("no", "csr","cython"))
worker.get_config()
```

```python
worker.process(img, variance=-img)
```
